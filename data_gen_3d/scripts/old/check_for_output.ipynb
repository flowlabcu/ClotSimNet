```python
import os

def find_output_dir(root_dir):
    for dirpath, dirnames, filenames in os.walk(root_dir):
        if 'output' in dirnames:
            print(f"'output' directory found at: {os.path.join(dirpath, 'output')}")
            return True
    print("'output' directory not found.")
    return False

# Example usage
find_output_dir('/mnt/ssd1/joshgregory/clotsimnet_data_3d')
```
